```python
X=np.vstack([X[data["target"]==i] for i in range(2)])
y=np.hstack([data["target"][data["target"]==i] for i in range(2)])
```

```python
import umap
c9_proj=umap.UMAP(n_neighbors=200,min_dist=0.1,metric='correlation').fit_transform(X)
```

```python
Umap=umap.UMAP(n_neighbors=9,n_components=3,metric='correlation',n_epochs=200,min_dist=0.01,verbose=True)
tsne=Umap.fit_transform(X)
```

```python
import seaborn as sns
# c9_proj=TSNE(random_state=RS,n_components=2,verbose=2,n_iter=1000,method='barnes_hut',init='pca').fit_transform(X)
def scatter(x,colors):
    palette=np.array(sns.color_palette("hls",9))
    f=plt.figure(figsize=(8,8))
    ax=plt.subplot(aspect='equal')
    sc=ax.scatter(x[:,0],x[:,1],lw=0,s=40,c=palette[colors.astype(np.int)])
    plt.xlim(-25,25)
    plt.ylim(-25,25)
    ax.axis('off')
    ax.axis('tight')
    txts=[]
    for i in range(9):
        xtext,ytext=np.median(x[colors==i,:],axis=0)
        txt=ax.text(xtext,ytext,c9.classes[i],fontsize=14)
        txt.set_path_effects([PathEffects.Stroke(linewidth=5,foreground='w'),PathEffects.Normal()])
        txts.append(txt)
    return f,ax,sc,txts
scatter(c9_proj,y)
```
